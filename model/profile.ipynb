# 문장 만들기
- 'doc2'를 분석하면 됨

```python
# 모든 문서의 공통 문장 제거
import re
def select(x):
  sen = re.split(r'[.?!\\\n]', x)
  result = ''
  for s in sen:
    if '어떤 맛을' in s:
      continue
    elif '어떤 음식과' in s:
      continue
    elif '알고 마시면 더' in s:
      continue
    else:
      if s != '':
        result += s + ". "
  return result

item_info['doc'] = item_info.detail.apply(select)
```

```python
# 분석 문장 생성
item_info['doc2'] = item_info.t_point + '. ' + item_info.s_point + '. ' + item_info.doc
item_info.head(3)
```

# item_profile 생성하기

```python
# 인덱스 순으로 토큰 리스트 가져오기
al_col = pd.Series(count_vect.vocabulary_).reset_index()
al_col = al_col.sort_values(by=0).set_index(0)
al_col = al_col['index'].to_list()
al_col
```

```python
term_doc = pd.DataFrame(feat_vect.toarray(), columns = al_col)
term_doc.head(3)

# 결측값 없음  =>  imputer 필요 없음
```

### 0~1점으로 minmaxscaler
->  각 값들의 최대 값을 1로, 작은 값을 0으로 정규화함  
->  아이템 토큰 지수 * 아이템 평가 점수의 평균으로 점수,,

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(term_doc[al_col])
item_profile = scaler.transform(term_doc[al_col])
item_profile = pd.DataFrame(item_profile, columns = al_col)
item_profile['al_id'] = item_info.id
item_profile['al_name'] = item_info.alname
item_profile.to_csv('item_profile.csv', encoding='utf8')
item_profile
```

# user_profile 생성

```python
users = pd.read_csv('/content/users.csv').drop(columns='Unnamed: 0')
users.head(2)
```

```python
reviews = pd.read_csv('/content/reviews.csv').drop(columns='Unnamed: 0')
reviews.head(3)
```

```python
# user_profile의 토큰 별 평균 점수 구하기
def tendency(x):
  # 이 함수 사용하면,, 평점 없는 술은 없다는 전제가 깔림,,

  # 사용자가 마신 술 아이디 리스트
  al_id_lst = reviews.loc[reviews.u_id == x].al_id.to_list()
  # 술에 평가한 스코어
  al_score_lst = reviews.loc[reviews.u_id == x].score.to_list()
  # 사용자가 구매한 아이템 수
  pur_num = users.loc[users.u_id == x].purchase_num

  # 중복 구매하는 아이템이 있음,, for문을 돌려보자
  # 토큰 분포
  token_dist = pd.DataFrame(columns = al_col)
  for idx, (al_id, al_score) in enumerate(zip(al_id_lst, al_score_lst)):
    a = item_profile.loc[item_profile.al_id == al_id][al_col] * al_score
    token_dist = token_dist.append(a, ignore_index=True)

  return token_dist.sum(axis = 0)/pur_num.values

  

demo = users[['u_id', 'u_name']]
demo[al_col] = demo.u_id.apply(tendency)
```

# 아이템 평가가 반영 되지 않은 사용자 리스트

```python
demo['total'] = demo[al_col].sum(axis=1)
# 리뷰 없는 사용자 아이디 리스트
no_reviewer_uid = demo.loc[demo.total == 0].u_id.to_list()
demo.loc[demo.total == 0]
```

```python
# 리뷰 없는 사용자 제외
user_profile = demo.loc[~demo.u_id.isin(no_reviewer_uid)][['u_id', 'u_name'] + al_col]
user_profile.to_csv('user_profile.csv', encoding='utf8')
user_profile
```

### lda로 문서의 토픽들을 뽑아보자

```python
# 카테고리 수를 21으로 임의 지정하여 나눠보자,,
lda = LatentDirichletAllocation(n_components=10, random_state=0)
lda.fit(feat_vect)
```

```python
# 21개의 토픽 별 주제 단어 비율
lda.components_
```

```python
# 350개의 피처 리스트
count_vect.get_feature_names()
```

```python
# 주제 별 주요 단어 보기

# no_top_words : 몇 개의 단어를 들고 올 것인가
# molde  :  lda가 되겠지,,
def display_topics(model, feature_names, no_top_words):
    for topic_index, topic in enumerate(model.components_):
        print('Topic #',topic_index)

        # components_ array에서 가장 값이 큰 순으로 정렬했을 때, 그 값의 array index를 반환. 
        topic_word_indexes = topic.argsort()[::-1]
        # 상위 인덱스만 가져오자  :  단어 수 만큼만..?
        top_indexes = topic_word_indexes[:no_top_words]
        
        # top_indexes대상인 index별로 feature_names에 해당하는 word feature 추출 후 join으로 concat
        feature_concat = ' '.join([feature_names[i] for i in top_indexes])                
        print(feature_concat)

# CountVectorizer객체내의 전체 word들의 명칭을 get_features_names( )를 통해 추출
feature_names = count_vect.get_feature_names()

# 토픽 대표 언어  :  문서별 대표 단어 아닌가봐
# Topic별 가장 연관도가 높은 word를 15개만 추출
display_topics(lda, feature_names, 15)
```
